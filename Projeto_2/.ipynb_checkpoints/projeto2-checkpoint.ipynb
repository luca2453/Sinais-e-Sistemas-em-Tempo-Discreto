# Funções utilizadas.

Coloque abaixo as funções desenvolvidas para este projeto.

```python
def iter_response(A, B, x):
    y = np.zeros(len(x))
    for i in range(0, len(x)):
        for j, a in enumerate(A[1:], start=1):
            if j<=i:
                y[i] -= a * y[i-j]
        for j, b in enumerate(B):
            if j<=i:
                y[i] += b * x[i-j]
        y[i] /= A[0]
    return y
```

# Questão 1

```python
fs, fc, lam = 500, 60, 0.9

wc = fc * 2 * np.pi / fs

print(f'Frequência de Corte: {wc}')

# Filtro NOTCH
z1 = np.exp(1j * wc)
z2 = np.exp(-1j * wc)

print(f'Zeros:')
print(f'Z1 = {z1}')
print(f'Z1 = {z2}')

b0 = np.real(1)
b1 = -np.real(z1 + z2)
b2 = np.real(z1 * z2)

B = np.array([b0, b1, b2])

p1 = lam * z1
p2 = lam * z2

print(f'Pólos:')
print(f'P1 = {p1}')
print(f'P1 = {p2}')

a0 = np.real(1)
a1 = -np.real(p1 + p2)
a2 = np.real(p1 * p2)

A = np.array([a0, a1, a2])

print(f'Numerador  : {A}')
print(f'Denominador: {B}')


sinal = sp.io.loadmat('noisy_ecg_data.mat')['noisy_ecg'].ravel()

resposta = iter_response(A, B, sinal)
t = np.linspace(0, len(resposta)/500, len(resposta))
plt.plot(t, sinal)
plt.plot(t, resposta)
```

# Questão 2

```python
fs = 500
fc = 35
l = 0.8

wc = fc * 2 * np.pi / fs

# low pass
p1 = l * np.exp( 1j * wc)
p2 = l * np.exp(-1j * wc)

print(f'Pólos:')
print(f'P1 = {p1}')
print(f'P1 = {p2}')

a0 = np.real(1)
a1 = -np.real(p1 + p2)
a2 = np.real(p1 * p2)

#Zeros
z1 = -1
z2 = -1

print(f'Zeros:')
print(f'Z1 = {z1}')
print(f'Z1 = {z2}')

b0 = np.real(1 )
b1 = -np.real(z1 + z2)
b2 = np.real(z1 * z2)

B = np.array([b0, b1, b2])
A = np.array([a0, a1, a2])

print(f'Numerador  : {A}')
print(f'Denominador: {B}')


#Calculando a resposta em frequência
w = np.linspace(-np.pi, np.pi, 1000)
z_axis = np.exp(1j * w)

H = (b0 * z_axis * z_axis + b1 * z_axis + b2) / (a0 * z_axis * z_axis + a1 * z_axis + a2)
H_mag = 10 * np.log10(abs(H))
H_phase = np.angle(H)

plt.plot(w, H_mag)
plt.title('Resposta em Frequência de H(e^jw)')
plt.xlabel('Frequência angular $\omega$')
plt.ylabel('Magnitude (dB)')
plt.show()

plt.plot(w, H_phase)
plt.title('Resposta em Frequência de H(e^jw)')
plt.xlabel('Frequência angular $\omega$')
plt.ylabel('Fase (radianos)')
plt.show()

sinal = sp.io.loadmat('noisy_ecg_data_2.mat')['noisy_ecg2'].ravel()

resposta = iter_response(A, B, sinal)
t = np.linspace(0, len(resposta)/500, len(resposta))
plt.plot(t, sinal)
plt.title('Sinal Original')
plt.show()

plt.plot(t, resposta)
plt.title('Sinal Filtrado')
plt.show()
```

# Questão 3

```python


# 1 - Considere o filtro Ha(s) = wc / s + wc (butterworth ordem 1).
w = np.linspace(0, 10**5*np.pi, 100)
wc = 10e3; 
Ha = wc / np.sqrt(w**2 + wc**2);

# 2 - Considere a aproximacao H1(z) com criterio de 10%
T = 10e-6*np.pi;
Hz = 0.3142 / np.sqrt(1.5333 - 1.4608 * np.cos(w*T));
# Hz = wc * T / np.sqrt(1 + np.exp(-2 * wc * T) - 2 * np.exp(-wc * T) * np.cos(w*T));

# 3 - Considere a aproximacao H2(z) com criterio de 1%.
T2 = 10e-7*np.pi;
Hz2 = wc * T2 / np.sqrt(1 + np.exp(-2 * wc * T2) - 2 * np.exp(-wc * T2) * np.cos(w*T2));

# figure;
# hold on
plt.plot(w, Ha,  'b-', linewidth=2, label='Filtro Analógico Ha(s)')
plt.plot(w, Hz,  'k-', linewidth=2, label='Filtro Digital H1(z) com Critério de 10%')
plt.plot(w, Hz2, 'r-', linewidth=2, label='Filtro Digital H2(z) com Critério de 1%')
plt.xlabel('Frequência angular $\omega$')
plt.ylabel('Magnitude')
plt.title('Aproximação de Filtros Analógicos')
plt.legend()
```


## Comparacao das respostas ao impulso


```python

tmax = 0.0001;
t = np.arange(0, tmax, 1e-6)
ha = wc * np.exp(-wc*t);

n1 = np.arange(0, tmax/T, 1)
h1 = wc * T * np.exp(-wc*T*n1);

n2 = np.arange(0, tmax/T2, 1)
h2 = wc * T2 * np.exp(-wc*T2*n2);


plt.plot(t,ha,'b-', label='h(s)')



markers, stems, base = plt.stem(n1*T,  h1/T,  'k', label='h2[n] com Critério de 10%')
plt.setp(markers, markersize=8)
markers2, stems2, base2 = plt.stem(n2*T2, h2/T2, 'r', label='h1[n] com Critério de 1%')
plt.setp(markers2, markersize=4)

plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.title('Aproximação de Filtros Analógicos - Resposta ao Impulso')
plt.legend()
```
